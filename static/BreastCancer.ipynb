```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from keras.models import Sequential
from keras.layers import Dense
```

```python
df = pd.read_csv("Breast_cancer.csv")

# For ease in demonstrational purposes we are only taking input of 4 parameters to predict the data(more attributes can be taken into consideration for final version of the product)
X = df[['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean']]
y = df['diagnosis']

# Encoding the labels to numerical format
label_encoder = LabelEncoder()
y = label_encoder.fit_transform(y)
```

```python

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```

```python

# We are defining our own custom Deep Neural Net model architecture
model = Sequential()
model.add(Dense(16, activation='relu', input_shape=(4,)))  # Input layer with 4 features
model.add(Dense(8, activation='relu'))  # Hidden layer with 8 units
model.add(Dense(1, activation='sigmoid'))  # Output layer with sigmoid activation

# We are using binary crossentropy as the loss function for sgd
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X_train, y_train, epochs=100, batch_size=32)

# Evaluating the model
loss, accuracy = model.evaluate(X_test, y_test)
print("Test loss:", loss)
print("Test accuracy:", accuracy)
```
